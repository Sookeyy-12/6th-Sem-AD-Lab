## Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Load Dataset

```python
df = pd.read_csv("../Data/japan_life_expectancy.csv")
df.head()
```

### Create Correlation table of all columns with Life Expectancy

```python
correlation_matrix = df.corr()
life_expectancy_corr = correlation_matrix["Life_expectancy"].sort_values(ascending=False)
print(life_expectancy_corr)
```

### Draw Upper and Lower Triagular Heatmap for the dataset

```python
# Full correlation matrix heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Full Correlation Matrix')
plt.show()

# Create upper triangle mask
upper_triangle = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))

# Upper triangle heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(upper_triangle, annot=True, cmap='coolwarm', fmt='.2f', mask=upper_triangle.isnull())
plt.title('Upper Triangle of Correlation Matrix')
plt.show()

# Create lower triangle mask
lower_triangle = correlation_matrix.where(np.tril(np.ones(correlation_matrix.shape), k=-1).astype(bool))

# Lower triangle heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(lower_triangle, annot=True, cmap='coolwarm', fmt='.2f', mask=lower_triangle.isnull())
plt.title('Lower Triangle of Correlation Matrix')
plt.show()
```

### Draw the 3 Different types of Correlation Matrix

```python
# Pearson correlation
pearson_corr = df.corr(method='pearson')
plt.figure(figsize=(12, 10))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Pearson Correlation Matrix')
plt.show()

# Spearman correlation
spearman_corr = df.corr(method='spearman')
plt.figure(figsize=(12, 10))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Spearman Correlation Matrix')
plt.show()

# Kendall correlation
kendall_corr = df.corr(method='kendall')
plt.figure(figsize=(12, 10))
sns.heatmap(kendall_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Kendall Correlation Matrix')
plt.show()
```

### Find the Best Correlation Matrix

1. Pearson Correlation:
    - You have continuous data that is normally distributed and linearly realted.

2. Spearman Correlation:
    - You have ordinal data or non-normally distributed continuous data that might have a monotonic relationship.

3. Kendall's Tau:
    - You have ordinal data or when you have a small sample size or data ties (duplicate ranks)
